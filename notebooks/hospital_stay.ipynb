# Preprocessing

```python
import pandas as pd

# Load the dataset (adjust path to your downloaded CSV file)
df = pd.read_csv("LengthOfStay.csv")
```

```python
# Convert 'vdate' to datetime
df['vdate'] = pd.to_datetime(df['vdate'])
```

```python
# Check unique values in categorical columns
print("Unique genders:", df['gender'].unique())
print("Unique discharged values:", df['discharged'].unique())
print("Unique facid values (facility IDs):", df['facid'].nunique())

# Convert categorical columns to category dtype or encode
df['gender'] = df['gender'].astype('category')
df['facid'] = df['facid'].astype('category')
```

## EDA

```python
# Describe target variable lengthofstay
print(df['lengthofstay'].describe())

# Plot distribution of lengthofstay
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['lengthofstay'], bins=30, kde=True)
plt.title("Distribution of Length of Hospital Stay")
plt.xlabel("Length of Stay (days)")
plt.show()
```

```python
# Convert discharged to datetime
df['discharged'] = pd.to_datetime(df['discharged'].astype(str))

# length of stay calc
df['calc_lengthofstay'] = (df['discharged'] - df['vdate']).dt.days

# Check if calc_lengthofstay matches lengthofstay
print((df['calc_lengthofstay'] == df['lengthofstay']).all())  # Should ideally be True

# one hot encoding
df = pd.get_dummies(df, columns=['facid'], drop_first=True)

# Gender encoding (Label encoding example)
df['gender'] = df['gender'].map({'F': 0, 'M': 1})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['lengthofstay'], bins=30)
plt.title("Distribution of Length of Stay")
plt.show()
```

## Feature Engineering

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score, KFold
import numpy as np

# Define columns to drop
leakage_cols = ['calc_lengthofstay']  # Add any other suspect columns here
cols_to_drop = leakage_cols + ['lengthofstay', 'eid'] # 'discharged' is handled below

# Separate target and features
X = df.drop(columns=cols_to_drop)
y = df['lengthofstay']

# Identify columns by type

datetime_cols = X.select_dtypes(include=['datetime64[ns]']).columns
num_cols = X.select_dtypes(include=['float64', 'int64']).columns.difference(datetime_cols)
cat_cols = X.select_dtypes(include=['category', 'object']).columns

# 1. Handle datetime columns: extract as ints

if not datetime_cols.empty:
    for col in datetime_cols:
        X[f'{col}_year'] = X[col].dt.year
        X[f'{col}_month'] = X[col].dt.month
        X[f'{col}_day'] = X[col].dt.day
    X = X.drop(columns=datetime_cols)  # drop original datetime columns after extraction

# Re-identify numerical columns after adding datetime features

num_cols = X.select_dtypes(include=['float64', 'int64']).columns

# 2. Impute numerical columns with median

if not num_cols.empty:
    X[num_cols] = X[num_cols].fillna(X[num_cols].median())

# 3. Impute categorical columns with mode
# Apply this only if cat_cols is not empty
if not cat_cols.empty:
    for col in cat_cols:
        X[col] = X[col].fillna(X[col].mode()[0])

# 4. Encode categorical variables

if not cat_cols.empty:
    le = LabelEncoder()
    for col in cat_cols:
        X[col] = le.fit_transform(X[col])

# Check the dtypes of the prepared X DataFrame
print("Data types after preprocessing:")
print(X.dtypes)

```

## Model Training & Evaluation


```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestRegressor
import numpy as np

model = RandomForestRegressor(n_estimators=10, random_state=42, n_jobs=-1)
kf = KFold(n_splits=3, shuffle=True, random_state=42)

# Sample 2000 rows only for quick testing
X_sample = X.sample(n=2000, random_state=42)
y_sample = y.loc[X_sample.index]

scores = cross_val_score(model, X_sample, y_sample, scoring='r2', cv=kf)
print(f"R² scores: {scores}")
print(f"Mean R²: {np.mean(scores):.4f}")
```

```python
model.fit(X, y)
importances = model.feature_importances_
feature_importance = sorted(zip(X.columns, importances), key=lambda x: x[1], reverse=True)
for feature, score in feature_importance[:10]:
    print(f"{feature}: {score:.4f}")
```
